PROBAR LAS SIGUIENTES HIPOTESIS

La hipotesis nula es que las calificaciones promedio de los usuarios para las plataformas Xbox One y PC son las mismas.
La hipotesis alternativa  es que las calificaciones promedio de los usuarios para las plataformas Xbox One y PC son diferentes.  

```python
df_xone=ventas_b.query("platform=='XOne'")
df_xone_score=df_xone.groupby('year_of_release')['user_score'].mean()
df_xone_score
```

```python
df_pc=ventas_b.query("platform=='PC'")
df_pc_score=df_pc.groupby('year_of_release')['user_score'].mean()
df_pc_score
```

```python
stat, p = st.levene(df_xone_score, df_pc_score)
p
```

```python
alpha=.05
results = st.ttest_ind(df_xone_score, df_pc_score) 
print('valor p:', results.pvalue) 

if results.pvalue < alpha:
    print("Rechazamos la hipótesis nula")
else:
    print("No podemos rechazar la hipótesis nula")
```

Las calificaciones promedio para las plataformas XOne y PC son las mismas. Para probar la hipotesis se 
usaron varianzas iguales porque en la prueba de levene dio p>0.05 lo que indica que las varianzas son estadisticamente iguales.
