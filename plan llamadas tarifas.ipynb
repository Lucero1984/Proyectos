## Prueba las hipótesis estadísticas

La hipotesis nula es que no hay diferencias en los ingresos promedio de los dos planes de tarifa

```python
stat, p = st.levene(ingreso_promedio_surf, ingreso_promedio_ult)
p
```

```python
# Prueba las hipótesis
alpha=.05
results = st.ttest_ind(ingreso_promedio_surf, ingreso_promedio_ult, equal_var=False) 
print('valor p:', results.pvalue) 

if results.pvalue < alpha:
    print("Rechazamos la hipótesis nula")
else:
    print("No podemos rechazar la hipótesis nula")
```

<div class="alert alert-block alert-success">
<b>Comentario de Revisor v2</b> <a class="tocSkip"></a>

Muy bien al usar el test de levene. El resultado del t-test es correcto también!
</div>

La hipotesis nula es que: el ingreso promedio mensual de los usuarios del area de NY-NJ es igual al ingreso promedio mensual de las otras ciudades.
La hipotesis alternativa es que el ingreso promedio mensual de los usuarios del area de NY-NJ es diferente al ingreso promedio mensual de las otras ciudades
